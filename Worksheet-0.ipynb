**1. Flatten a Nested List:**

Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.


*   Requirements:

    – Define a function flatten(lst) that takes a nested list lst and returns a flattened version
    of the list.

    – Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].



```python
def flatten(lst):
    """
    Flatten a nested list into a single list.

    Parameters:
        lst (list): A nested list.

    Returns:
        list: A flattened list containing all elements.
    """

    flat_list = []

    # Go through each sublist
    for sublist in lst:
        # Go through each element inside the sublist
        for item in sublist:
            flat_list.append(item)

    return flat_list


# Ask the user to enter nested lists
user_input = input("Enter nested lists (example: [[1,2],[3,4],[5]]): ")

# Convert input text into a real Python list
nested_list = eval(user_input)

# Call the function
result = flatten(nested_list)

# Print the result
print("Flattened list:", result)
```

**3. Sum of All Elements in a Nested List:**

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).



*   Requirements:

    – Define a function sum nested(lst) that takes a nested list lst and returns the sum of all
    the elements.
    
    – Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21.



```python
def sum_nested(lst):
    """
    Calculate the sum of all numbers in a nested list.

    Parameters:
        lst (list): A nested list containing numbers.

    Returns:
        int/float: The total sum of all elements.
    """

    total = 0

    for item in lst:
        # If the item is a list, call the function again (recursion)
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item

    return total


# Ask the user to enter a nested list
user_input = input("Enter a nested list (example: [[1,2],[3,[4,5]],6]): ")

# Convert the input string to a Python list
nested_list = eval(user_input)

# Call the function
result = sum_nested(nested_list)

# Print the result
print("Sum of all elements:", result)
```

Numbers inside:

1 + 2 + 3 + 4 + 5 + 6

Result:

**4. Remove Specific Element from a Nested List:**

Write a Python function that removes all occurrences of a specific element from a nested list.
*   Requirements:

    – Define a function remove element(lst, elem) that removes elem from lst and returns the
    modified list.

    – Example: For the input lst = [[1, 2], [3, 2], [4, 5]] and elem = 2, the output should
    be [[1], [3], [4, 5]].





```python
def remove_element(lst, elem):
    """
    Remove all occurrences of a specific element from a nested list.

    Parameters:
        lst (list): A nested list.
        elem: The element to remove.

    Returns:
        list: Modified nested list without the element.
    """

    new_list = []

    for item in lst:
        # If the item is a list, apply the function recursively
        if isinstance(item, list):
            new_list.append(remove_element(item, elem))
        else:
            # Only add the item if it is not equal to elem
            if item != elem:
                new_list.append(item)

    return new_list


# Ask the user to enter a nested list
user_list = input("Enter a nested list (example: [[1,2],[3,2],[4,5]]): ")

# Convert input to Python list
nested_list = eval(user_list)

# Ask the user which element to remove
element = int(input("Enter the element to remove: "))

# Call the function
result = remove_element(nested_list, element)

# Print the result
print("List after removing element:", result)
```

Remove 2.

    – Define a function find max(lst) that takes a nested list lst and returns the maximum
    element.

    – Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    """
    Find the maximum element in a nested list.

    Parameters:
        lst (list): A nested list containing numbers.

    Returns:
        int/float: The maximum element found in the list.
    """

    max_value = float('-inf')  # start with a very small number

    for item in lst:
        # If the item is a list, search inside it
        if isinstance(item, list):
            value = find_max(item)
        else:
            value = item

        # Update the maximum value
        if value > max_value:
            max_value = value

    return max_value


# Ask the user to enter a nested list
user_input = input("Enter a nested list (example: [[1,2],[3,[4,5]],6]): ")

# Convert input string into a Python list
nested_list = eval(user_input)

# Call the function
result = find_max(nested_list)

# Print the result
print("Maximum element:", result)
```

**6. Count Occurrences of an Element in a Nested List:**

Write a Python function that counts how many times a specific element appears in a nested list.

* Requirements:

    – Define a function count occurrences(lst, elem) that counts the occurrences of elem in
    the nested list lst.

    – Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should
    be 3.

```python
def count_occurrences(lst, elem):
    """
    Count how many times a specific element appears in a nested list.

    Parameters:
        lst (list): A nested list.
        elem: The element to count.

    Returns:
        int: Number of times elem appears in the list.
    """

    count = 0

    for item in lst:
        # If the item is another list, search inside it
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1

    return count


# Ask the user to enter a nested list
user_list = input("Enter a nested list (example: [[1,2],[2,3],[2,4]]): ")

# Convert input text to a Python list
nested_list = eval(user_list)

# Ask the user which element to count
element = int(input("Enter the element to count: "))

# Call the function
result = count_occurrences(nested_list, element)

# Print the result
print("Number of occurrences:", result)
```

**7. Flatten a List of Lists of Lists:**

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

* Requirements:

    – Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single
    flattened list.

    – Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
    [1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
    """
    Flatten a deeply nested list into a single list.

    Parameters:
        lst (list): A deeply nested list.

    Returns:
        list: A flattened list containing all elements.
    """

    flat_list = []

    for item in lst:
        # If the item is another list, flatten it recursively
        if isinstance(item, list):
            flat_list.extend(deep_flatten(item))
        else:
            flat_list.append(item)

    return flat_list


# Ask the user to enter a deeply nested list
user_input = input("Enter a nested list (example: [[[1,2],[3,4]],[[5,6],[7,8]]]): ")

# Convert input string to Python list
nested_list = eval(user_input)

# Call the function
result = deep_flatten(nested_list)

# Print the result
print("Flattened list:", result)
```

Simple Explanation

Example list:

[[[1,2],[3,4]],[[5,6],[7,8]]]

Numbers inside:

1,2,3,4,5,6,7,8

Final flattened list:

**8. Nested List Average:**

Write a Python function that calculates the average of all elements in a nested list.

* Requirements:

    – Define a function average nested(lst) that takes a nested list lst and returns the average
    of all the elements.

    – Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    """
    Calculate the average of all numbers in a nested list.

    Parameters:
        lst (list): A nested list containing numbers.

    Returns:
        float: The average of all elements.
    """

    total = 0
    count = 0

    for item in lst:
        # If item is another list, process it recursively
        if isinstance(item, list):
            sub_total, sub_count = average_nested_helper(item)
            total += sub_total
            count += sub_count
        else:
            total += item
            count += 1

    return total / count


def average_nested_helper(lst):
    """
    Helper function to calculate total and count of nested list elements.
    """

    total = 0
    count = 0

    for item in lst:
        if isinstance(item, list):
            sub_total, sub_count = average_nested_helper(item)
            total += sub_total
            count += sub_count
        else:
            total += item
            count += 1

    return total, count


# Ask the user to enter a nested list
user_input = input("Enter a nested list (example: [[1,2],[3,4],[5,6]]): ")

# Convert the input string into a Python list
nested_list = eval(user_input)

# Call the function
result = average_nested(nested_list)

# Print the result
print("Average of all elements:", result)
```

Simple Explanation

Example list:

[[1,2],[3,4],[5,6]]

Numbers:

1,2,3,4,5,6

Total:

1+2+3+4+5+6 = 21

Count:

6 numbers

Average:

21 ÷ 6 = 3.5

3. Prove (AB)T = BTAT

```python
import numpy as np  # import numpy library for matrix operations

# define matrices
A = np.array([[3,4],[7,8]])  # matrix A
B = np.array([[5,3],[2,1]])  # matrix B

# print matrices
print("Matrix A:")
print(A)

print("Matrix B:")
print(B)

# compute AB
AB = np.dot(A, B)  # matrix multiplication

# transpose of AB
AB_T = AB.T  # .T gives transpose of a matrix

# transpose of A and B
A_T = A.T
B_T = B.T

# compute B^T * A^T
BT_AT = np.dot(B_T, A_T)

# print results
print("Transpose of (AB):")
print(AB_T)

print("B^T * A^T:")
print(BT_AT)

# check if they are equal
print("Are they equal?")
print(np.array_equal(AB_T, BT_AT))
```

3x + y − z = 9
